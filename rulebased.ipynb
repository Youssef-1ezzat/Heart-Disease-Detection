```python
from experta import *

class HeartDiseaseExpert(KnowledgeEngine):

    @Rule(Fact(age=P(lambda x: x > 50)), Fact(chol=P(lambda x: x > 240)))
    def high_risk1(self):
        print("High risk of heart disease")

    @Rule(Fact(trestbps=P(lambda x: x > 140)))
    def high_bp(self):
        print("High blood pressure risk")

    @Rule(Fact(exang=1))
    def exercise_angina(self):
        print("Chest pain during exercise")

    @Rule(Fact(oldpeak=P(lambda x: x > 2)))
    def st_depression(self):
        print("Possible heart disease")

    @Rule(Fact(thalach=P(lambda x: x < 100)))
    def low_heart_rate(self):
        print("Low heart rate risk")

    @Rule(Fact(cp=3))
    def chest_pain(self):
        print("Severe chest pain")

    @Rule(Fact(ca=P(lambda x: x > 2)))
    def artery_block(self):
        print("Possible artery blockage")

    @Rule(Fact(fbs=1))
    def high_sugar(self):
        print("High fasting blood sugar")

    @Rule(Fact(age=P(lambda x: x > 60)))
    def old_age(self):
        print("Age risk factor")

    @Rule(Fact(chol=P(lambda x: x > 300)))
    def very_high_chol(self):
        print("Very high cholesterol")
```

```python
engine = HeartDiseaseExpert()
engine.reset()

engine.declare(Fact(age=55))
engine.declare(Fact(chol=250))

engine.run()
```
